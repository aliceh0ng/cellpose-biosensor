Save stitched mask as colored PNG for methods figure

```python
import numpy as np
import tifffile
from pathlib import Path
from PIL import Image

mask_path = Path('/Users/alicehong/projects/cellpose-biosensor/data/processed/overlays/Scene-01-20260416-C3M2_Tcol_1-A01Export-01_c1-4_stack/mask_stitched.tif')
png_path = mask_path.with_name('mask_stitched_colored.png')

mask = tifffile.imread(mask_path)
labels = np.unique(mask)
labels = labels[labels != 0]

# Black background for label 0; bright deterministic colors for each cell label.
rng = np.random.default_rng(42)
lut = np.zeros((int(mask.max()) + 1, 3), dtype=np.uint8)
lut[labels] = rng.integers(60, 256, size=(len(labels), 3), dtype=np.uint8)

rgb = lut[mask.astype(np.int64)]
Image.fromarray(rgb, mode='RGB').save(png_path)

print(f'Saved colored mask PNG -> {png_path}')
print(f'{len(labels)} labeled masks')
```
